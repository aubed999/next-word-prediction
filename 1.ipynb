```python
def getlines(path):
    data=[]
    with open(path, "r", encoding='utf-8') as lines:
        for line in lines:
            temp = line.strip()
            temp = temp.replace("!","")
            temp = temp.replace("?","")
            temp = temp.replace(".","")
            temp = temp.replace("،","")   
            data.append(temp)     
    return data
sent = getlines("train.txt")
print(sent[0])
```

```python
def cal_unigram(corpus):
    count_words = {}
    total = 0

    for line in corpus:
        tmp = line.split(" ")
        total += len(tmp)

        for word in tmp:
            if not(word in count_words):
                count_words[word]=0
            count_words[word] +=1
    return count_words, total

count_words, total = cal_unigram(sent)

s = dict(sorted(count_words.items(), key=lambda item:item[1], reverse=True)[:20])
#print(s)
c=1
for key,value in s.items():
    print(key,value,value*c)
    c+=1
```

```python
def cal_bigram(corpus):
    
    bigram_words={}
    for line in corpus:
        temp = line.split(" ")
        for i in range(1, len(temp)):
            word = temp[i]
            pre_word = temp[i-1]

            if not (word in bigram_words):
                bigram_words[word]=[]
            bigram_words[word].append(pre_word)
    return bigram_words

dic_bigrm=cal_bigram(sent)
print(dic_bigrm["محکم"])
len(dic_bigrm["محکم"])
```

```python
def bigram_prob(word, prev_word):
    

    count = 0
    for i in dic_bigrm[word]:
        if i == prev_word:
            count+=1
    number = count_words[prev_word]

    prob = count/number

    return prob
b_prob= bigram_prob("محکم","تو")
print(b_prob)
```

```python
def bigram_sequence_prob(sentence):
    
    prob = 1
    temp = sentence.split(" ")
    prob = prob * (count_words[temp[0]]/total)

    for i in range(1, len(temp)):
        count = 0
        for j in dic_bigrm[temp[i]]:
            if j == temp[i-1]:
                count+=1
        prob = prob * bigram_prob(temp[i],temp[i-1])

    return prob
print(bigram_sequence_prob("شب است و شمع و شراب و شیرینی"))
```

```python
def cal_trigram(corpus,word):
    
    trigram_words=[]
    for line in corpus:
        temp = line.split(" ")
        for i in range(2, len(temp)):
            sequence = []
            if word == temp[i]:
                sequence.append(temp[i-1])
                sequence.append(temp[i-2])
            if len(sequence) == 2:   
                trigram_words.append(sequence)
    return trigram_words
cal_trigram(sent,"محکم")
```

```python
def trigram_prob(word, prev_word1, prev_word2):
    
    sequence = []
    sequence.append(prev_word1)
    sequence.append(prev_word2)
    countW=0
    for i in cal_trigram(sent,word):
        if i == sequence:
           countW+=1 

    number=1
    for j in dic_bigrm[prev_word1]:
        if j == prev_word2:
            number+=1
    
    t_prob = countW/number
    return t_prob

trigram_prob("حکم","محکم","از") 
```

```python
def trigram_sequence_prob(sentence):
    prob = 1.0
    temp = sentence.split(" ")
    prob = prob * (count_words[temp[0]]/total)
    b_prob = bigram_prob(temp[1],temp[0])
    prob = prob * (b_prob)

    for i in range(2, len(temp)):
        prob = prob * trigram_prob(temp[i],temp[i-1],temp[i-2])

    return prob
print(trigram_sequence_prob("شب است و شمع و شراب و شیرینی"))
```
